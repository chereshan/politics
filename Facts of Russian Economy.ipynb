```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fedbud_year = pd.read_excel(r'C:\Users\79618\Downloads\fedbud_year.xlsx',header=0)
```

```python
fedbud_year
```

```python
plt.plot(12*40245*145000000/(fedbud_year.iloc[13,5:]*1000000000)*100,label='40.245, медиана, 2021')
plt.plot(12*30000*145000000/(fedbud_year.iloc[13,5:]*1000000000)*100,label='30.000')
plt.plot(12*25000*145000000/(fedbud_year.iloc[13,5:]*1000000000)*100,label='25.000')
plt.plot(12*20000*145000000/(fedbud_year.iloc[13,5:]*1000000000)*100,label='20.000')
plt.plot(12*15000*145000000/(fedbud_year.iloc[13,5:]*1000000000)*100,label='15.000')
plt.plot(12*10000*145000000/(fedbud_year.iloc[13,5:]*1000000000)*100,label='10.000')
plt.legend()
plt.title('Какой процент от федеральных госрасходов составит БОД')
plt.xlabel('Год')
plt.ylabel('Доля от расходов бюджета, %')
plt.grid()
```
